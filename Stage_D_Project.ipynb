```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import altair as alt
p_data = pd.read_csv("/content/africa_food_production.csv")
s_data = pd.read_csv('/content/Africa_Food_Supply.csv')
```

```python
plt.figure(figsize=(20,10))
sns.barplot(y = 'Country', x = 'Value', data = p_data,errorbar = None, color = 'red')                     #data of total production
```

```python
m1 = p_data.groupby('Country')['Value'].mean()
m1
```

```python
plt.figure(figsize=(20,10))
sns.barplot(data=m1,errorbar = None, color = 'red')                                          #plot of means of country's production
plt.xticks(rotation=90)
```

```python
m1 = pd.DataFrame(data = m1)
m1 = p_data.groupby('Country')['Value'].mean().reset_index()
m1['Country'] = pd.Categorical(m1['Country'])                                #boxplot on production data
m1['Country'] = m1['Country'].cat.codes
plt.boxplot(x = 'Value',data = m1, widths = 800)
plt.show()
```
